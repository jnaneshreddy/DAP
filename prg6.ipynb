```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x=np.array([2,4,6,8,9])
y=np.array([1,3,5,7,10])
```

```python
plt.figure(figsize=(6,4))
plt.plot(x,y)
plt.ylabel("y-axis")
plt.xlabel("x-axis")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(x,y)
plt.title("scatter plot")
plt.xlabel("x-axis")
plt.ylabel("yaxis")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.bar(x,y)
plt.title("bar graph")
plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.pie(x,labels=['a','b','c','d','e'])
plt.title("pie graph")
plt.ylabel("y-axis")
plt.xlabel("x-axis")
plt.show()
```
