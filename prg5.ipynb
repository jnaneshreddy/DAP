```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr1 = np.array([[1,2,3],[4,5,6]])
arr2 = np.array([[7,8,9],[10,11,12]])
```

```python
concatinanted_arr = np.concatenate((arr1,arr2))
```

```python
reshaped_arr = np.reshape(concatinanted_arr,(3,4))
print("array manipulation")
print("original array")
print(arr1)
print(arr2)
print("concatinated arry")
print(concatinanted_arr)
print("reshaped arrray")
print(reshaped_arr)
```

```python
sorted_arr = np.sort(reshaped_arr)
print("sorting  sorted array")
print(sorted_arr)
```

```python
splited_arr = np.split(sorted_arr,3)
print("splitting")
print("splitted array")
for i,arr in enumerate(splited_arr):
    print(f"array{i+1} :{arr}")
```

```python
brodcasted_arr = reshaped_arr+10
print("brodcasting...")
print("original array")
print(reshaped_arr)
print("brodcasted array")
print(brodcasted_arr)
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)

plt.plot(x,y)
plt.title("sin x")
plt.xlabel("x Axis")
plt.ylabel("y-axis")
plt.show()
```
