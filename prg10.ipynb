```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
date_range = pd.date_range(start='2022-01-01',end='2022-01-10',freq='D')
data = {'value':[10,20,18,34,16,17,33,25,66,80]}
df=pd.DataFrame(data,index=date_range)
```

```python
print("time series data")
print(df)
```

```python
plt.figure(figsize=(15,4))
plt.plot(df.index,df['value'],marker='o',linestyle ='-',color='b')
plt.title('Time series  Analysis')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python
print("basic time series oprations")
print("minimum valu:",df['value'].min())
print("maximum value: ",df['value'].max())
print("mean  value",df['value'].mean())
```

```python
weekly_reshamped = df.resample('W').mean()
print("\n Resampled array weekly frequancy  :")
print(weekly_reshamped)
```
